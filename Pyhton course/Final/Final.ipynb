```python
"""Problem 1: Employee Salary Analysis"""

import pandas as pd
import os
data = {
    "name": ["MR.A", "MR.B", "MR.C", "MR.D", "MR.E", "MR.F", "MR.G", "MR.H"],
    "age": [25, 30, 35, 40, 29, 33, 27, 31],
    "salary": [50000, 60000, 70000, 80000, 55000, 65000, 72000, 58000],
    "department": ["HR", "IT", "IT", "Finance", "HR", "Finance", "IT", "HR"]
}
df = pd.DataFrame(data)
df.to_csv("employees_large.csv", index=False)
print("Sample employees_large.csv file created successfully!")

if "employees_large.csv" in os.listdir():
    print("File found! Proceeding with analysis.")
else:
    print("Error: File not found!")

df_employees = pd.read_csv('employees_large.csv')

average_salary = df_employees.groupby('department')['salary'].mean()
print("Average Salary by Department:\n", average_salary)

highest_salary_department = average_salary.idxmax()
print("Department with Highest Average Salary:", highest_salary_department)
```

```python
"""Problem 2: Sales Data Analysis"""
import pandas as pd
import os


if "sales_large.csv" not in os.listdir():
    sales_data = {
        "region": ["North", "South", "East", "West", "North", "South", "East", "West"],
        "product": ["A", "B", "C", "A", "B", "C", "A", "B"],
        "sales": [1000, 1500, 1200, 1700, 1800, 1600, 1400, 1300]
    }

    df_sales = pd.DataFrame(sales_data)
    df_sales.to_csv("sales_large.csv", index=False)
    print("\nSample sales_large.csv file created successfully!")

df_sales = pd.read_csv('sales_large.csv')

total_sales = df_sales.groupby(['region', 'product'])['sales'].sum()
print("\nTotal Sales by Region and Product:\n", total_sales)

top_product = df_sales.groupby('product')['sales'].sum().idxmax()
print("\nProduct with Highest Total Sales:", top_product)
```

```python
"""Problem 3: Matrix Operations with NumPy"""

import numpy as np

A = np.array([[2,4,6], [1,3,5], [7,9,11]])
B = np.array([[8,5, 2], [7,4,1], [6,3,0]])
sum_matrix = A + B
product_matrix= A * B
print("Sum of matrices:")
print(sum_matrix)
print("Product of matrices:")
print(product_matrix)

B_transpose = B.T
result = np.dot(A, B_transpose)
print("Result of A * B^T:")
print(result)
determinant = np.linalg.det(A)
print("Determinant of the matrix A:", determinant)
```

```python
"""Problem 4: Statistical Analysis of Random Dataset"""

data = np.random.randn(10000)

mean_val = np.mean(data)
median_val = np.median(data)
std_dev = np.std(data)
print(f"Mean: {mean_val}, Median: {median_val}, Standard Deviation: {std_dev}")

min_val, max_val = np.min(data), np.max(data)
print(f"Minimum: {min_val}, Maximum: {max_val}")

count_greater_mean = np.sum(data > mean_val)
print(f"Count of values greater than mean: {count_greater_mean}")

data_normalized = (data - min_val) / (max_val - min_val)
print("First 10 normalized values:", data_normalized[:10])
```

```python
"""Problem 5: List Comprehension with Temperature Data"""

temperatures = [28, 32, 35, 31, 29, 30, 33, 36, 27, 25, 34, 30, 29, 37, 38, 26, 31, 35, 33, 32, 36, 34, 29, 28, 27, 35, 32, 30, 31, 33]

above_32 = [temp for temp in temperatures if temp > 32]
print("Temperatures above 32°C:", above_32)

classified_temps = [(temp, "Hot" if temp > 35 else "Warm" if 30 <= temp <= 35 else "Cool") for temp in temperatures]
print("Classified Temperatures:", classified_temps)

avg_temp = sum(temperatures) / len(temperatures)
temp_differences = [temp - avg_temp for temp in temperatures]
print("Temperature Differences from Monthly Average:", temp_differences)
```
